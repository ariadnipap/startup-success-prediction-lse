# Week 5 Task 1: Compare all models

In this document metrics from those models trained before will be loaded:
- Logistic Regression
- Random Forest
- XGBoost

The metrics include:
- Accuracy
- Precision
- Recall
- F1 Score


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

metrics = {
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "XGBoost Baseline",
        "XGBoost Tuned"
    ],
    "Accuracy": [
        0.714,     # LR
        0.750,     # RF
        0.7568,    # XGB Baseline
        0.7405     # XGB Tuned
    ],
    "Precision": [
        0.782,     # LR
        0.790,     # RF
        0.7953,    # XGB Baseline
        0.7687     # XGB Tuned
    ],
    "Recall": [
        0.775,     # LR
        0.840,     # RF
        0.8417,    # XGB Baseline
        0.8583     # XGB Tuned
    ],
    "F1 Score": [
        0.778,     # LR
        0.810,     # RF
        0.8178,    # XGB Baseline
        0.8110     # XGB Tuned
    ]
}
```

```python
# Create a DataFrame from the metrics dictionary
df_metrics = pd.DataFrame(metrics)

# Optional: sort by F1 score descending
df_metrics_sorted = df_metrics.sort_values(by="F1 Score", ascending=False)

print(df_metrics)
print("\nSorted by F1 Score:")
print(df_metrics_sorted)

df_metrics.to_csv("..\docs\Week5_Four_model_comparison\comparison_metrics.csv", index=False)
print("Saved: comparison_metrics.csv")
```

```python
# plot bar charts
# df is df_metrics

metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
models = df_metrics["Model"].tolist()
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

# Create figure
plt.figure(figsize=(14, 10))

for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    values = df_metrics[metric].tolist()

    # Plot bar chart
    plt.bar(models, values, color=colors)

    # Metric-specific y-axis scaling
    ymin = min(values) - 0.02
    ymax = max(values) + 0.02
    plt.ylim(ymin, ymax)

    plt.title(metric)
    plt.ylabel("Value")
    plt.xticks(rotation=30, ha='right')

    plt.tight_layout()

# Save final figure
plt.savefig("..\docs\Week5_Four_model_comparison\model_comparison_plot.png", dpi=300, bbox_inches="tight")
plt.show()
print("Saved: model_comparison_plot.png")
```
